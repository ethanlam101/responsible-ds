```python
import pandas as pd
import numpy as np
```

```python
import zipfile

dataframes = []
with zipfile.ZipFile('hospitality_data.zip', 'r') as zip_ref:
    zip_ref.extractall()
for dates in [2022,2023,2024,2025]:
    df_1 = pd.read_csv(f"{dates}\HCAHPS-Hospital.csv", encoding='latin1', dtype={'Facility ID': str})
    df_2 = pd.read_csv(f"{dates}\Hospital_General_Information.csv", encoding='latin1', dtype={'Facility ID': str})
    df_3 = pd.read_csv(f"{dates}\OQR_OAS_CAHPS_BY_HOSPITAL.csv", encoding='latin1', dtype={'Facility ID': str})
    df_4 = pd.read_csv(f"{dates}\PCH_HCAHPS_HOSPITAL.csv", encoding='latin1', dtype={'Facility ID': str})

    main_df = pd.merge(df_1, df_2, how='inner', on='Facility ID')
    main_df = pd.merge(main_df, df_3, how='left', on='Facility ID')
    main_df = pd.merge(main_df, df_4, how='left', on='Facility ID')
    dataframes.append(main_df)
```

```python
yr2022_df = dataframes[0]
yr2023_df = dataframes[1]
yr2024_df = dataframes[2]
yr2025_df = dataframes[3]
```
